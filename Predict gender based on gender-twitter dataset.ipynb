# FEATURE SELECTION

```python
df.describe()
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['_unit_state']=le.fit_transform(df['_unit_state'])
df['_last_judgment_at']=le.fit_transform(df['_last_judgment_at'])
df['gender']=le.fit_transform(df['gender'])
df['profile_yn']=le.fit_transform(df['profile_yn'])
df['created']=le.fit_transform(df['created'])
df['description']=le.fit_transform(df['description'])

df['gender_gold']=le.fit_transform(df['gender_gold'])
df['link_color']=le.fit_transform(df['link_color'])
df['name']=le.fit_transform(df['name'])
df['text']=le.fit_transform(df['text'])
df['profileimage']=le.fit_transform(df['profileimage'])
df['sidebar_color']=le.fit_transform(df['sidebar_color'])
df['tweet_created']=le.fit_transform(df['tweet_created'])
df['tweet_location']=le.fit_transform(df['tweet_location'])
df['user_timezone']=le.fit_transform(df['user_timezone'])
                                   
```

```python
y=df[['gender']].values
'''
x=df[['user_timezone','tweet_id','tweet_count','sidebar_color','profileimage',
     'link_color','fav_number','created','profile_yn','_last_judgment_at',
     '_unit_state']].values  
     
     33.5% for svm
     45.76% for knn
     53.87% for random forest
     with feature selection from heat map(only 12 out of 23 cols->features)
     
     x=df[['_unit_id', '_unit_state', '_trusted_judgments', '_last_judgment_at',
       'gender:confidence', 'profile_yn', 'profile_yn:confidence',
       'created', 'description', 'fav_number', 'gender_gold', 'link_color',
       'name', 'profileimage', 'retweet_count', 'sidebar_color', 'text',
       'tweet_count', 'tweet_created', 'tweet_id', 'tweet_location',
       'user_timezone']].values
     
     On giving all the cols except gender as features:
     33.42% for svm
     45.72% for knn
     57.19% for random forest

'''

x=df[['user_timezone','tweet_id','tweet_count','sidebar_color','profileimage',
     'link_color','fav_number','created','profile_yn','_last_judgment_at',
     '_unit_state']].values

#gender depends only on 12 cols out of 23 cols
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

# SVM

```python
from sklearn.svm import SVC
```

```python
model1 = SVC()
```

```python
model1.fit(x_train,y_train)
```

```python
model1_predictions = model1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(model1_predictions,y_test)
```
